```python
"""
NATIONAL MATERNAL MORTALITY TREND 2018-2023
Shows overall national improvement following Medicaid extension
For Policy Brief - FIXED TEXT OVERLAP
"""

import matplotlib.pyplot as plt
import numpy as np

# National MMR data (all races combined)
years = ['2018', '2019', '2020', '2021', '2022', '2023']
mmr_national = [17.4, 20.1, 23.8, 32.9, 22.3, 18.2]

# Create figure
fig, ax = plt.subplots(figsize=(12, 7.5), facecolor='white')

# Main trend line
line = ax.plot(years, mmr_national, 
               marker='o', markersize=12, linewidth=3.5, 
               color='#003D82', markerfacecolor='#F4A03F', 
               markeredgewidth=2.5, markeredgecolor='#003D82',
               label='National MMR')

# Highlight key points
ax.scatter([3], [32.9], s=500, color='#DC143C', alpha=0.25, zorder=5)
ax.annotate('COVID Peak\n32.9 per 100K', 
            xy=(3, 32.9), xytext=(3, 37),
            ha='center', fontsize=11, fontweight='bold', color='#DC143C',
            arrowprops=dict(arrowstyle='->', color='#DC143C', lw=2))

ax.scatter([5], [18.2], s=500, color='#32CD32', alpha=0.25, zorder=5)
ax.annotate('Post-Extension\n18.2 per 100K', 
            xy=(5, 18.2), xytext=(5, 13),
            ha='center', fontsize=11, fontweight='bold', color='#32CD32',
            arrowprops=dict(arrowstyle='->', color='#32CD32', lw=2))

# Add improvement arrow
ax.annotate('', 
            xy=(5, 18.2), xycoords='data',
            xytext=(3, 32.9), textcoords='data',
            arrowprops=dict(arrowstyle='<->', color='#F4A03F', lw=3))

ax.text(4, 26, '45% Improvement\n(14.7 point decline)', 
        ha='center', fontsize=13, fontweight='bold', color='#F4A03F',
        bbox=dict(boxstyle='round,pad=0.7', facecolor='white', 
                 edgecolor='#F4A03F', linewidth=2))

# Add policy implementation line
ax.axvline(x=3.5, color='#008C95', linestyle='--', linewidth=3, alpha=0.7, 
           label='Medicaid Postpartum\nExtension (April 2022)')

# Add value labels on all points
for i, (year, mmr) in enumerate(zip(years, mmr_national)):
    if i not in [3, 5]:  # Skip the annotated points
        # SPECIAL HANDLING FOR 22.3 (index 4) - Move it down more
        if i == 4:
            ax.text(i, mmr - 2, f'{mmr}',  # MOVED DOWN (mmr - 2 instead of mmr + 1.5)
                    ha='center', va='top', fontsize=11, fontweight='bold', color='#2C3E50')
        else:
            ax.text(i, mmr + 1.5, f'{mmr}', 
                    ha='center', va='bottom', fontsize=11, fontweight='bold', color='#2C3E50')

# Labels and styling
ax.set_xlabel('Year', fontsize=14, fontweight='bold', color='#003D82')
ax.set_ylabel('Maternal Mortality Rate (per 100,000 live births)', fontsize=14, fontweight='bold', color='#003D82')
ax.set_title('National Maternal Mortality Trend: Dramatic Improvement Following Medicaid Postpartum Extension\nApproximately 550 Fewer Maternal Deaths Annually', 
             fontsize=15, fontweight='bold', color='#003D82', pad=20)

ax.set_ylim(0, 40)
ax.set_yticks(np.arange(0, 41, 5))
ax.yaxis.grid(True, alpha=0.3, linestyle='--', color='gray')
ax.legend(loc='upper left', fontsize=11, framealpha=0.9)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#2C3E50')
ax.spines['bottom'].set_color('#2C3E50')

# Spacing
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)

# Add data source note
fig.text(0.5, 0.06, 
         'Source: CDC WONDER Database, 2018-2023. Maternal mortality defined as deaths during pregnancy or within 42 days of pregnancy termination.',
         ha='center', fontsize=9, style='italic', color='#2C3E50')

plt.savefig('national_mmr_trend_2018_2023.png', dpi=300, bbox_inches='tight', facecolor='white')
print("âœ… Chart saved: national_mmr_trend_2018_2023.png")
plt.show()
```

```python
"""
VOLATILITY BY INFRASTRUCTURE SCORE CHART
For Policy Brief - CORRECTED VERSION WITH PROPER FOOTNOTE PLACEMENT

Infrastructure scored 0-4 based on presence of 4 key indicators:
- Medicaid Expansion (Process pillar)
- Doula Coverage (People pillar)  
- Paid Family Leave (People pillar)
- Established MMRC (Technology pillar)

States can score 0 (none) to 4 (all indicators present)
"""

import matplotlib.pyplot as plt
import numpy as np

# Data from your framework validation
infrastructure_scores = [0, 1, 2, 3, 4]

# Average Coefficient of Variation (%) for each infrastructure score
# From your analysis: higher score = lower volatility = more stable
volatility_cv = [33.6, 29.0, 27.6, 34.7, 15.2]

# Number of states in each category
state_counts = [5, 5, 6, 6, 7]

# Create figure with professional styling
fig, ax = plt.subplots(figsize=(11, 7), facecolor='white')  # Made taller for footnote

# Create bar chart with color gradient (red=bad, green=good)
colors = ['#DC143C', '#FF6347', '#FFA500', '#FFD700', '#32CD32']
bars = ax.bar(infrastructure_scores, volatility_cv, 
               color=colors,
               edgecolor='#2C3E50', linewidth=1.5, alpha=0.9)

# Add value labels on top of bars
for i, (score, cv, count) in enumerate(zip(infrastructure_scores, volatility_cv, state_counts)):
    # Percentage on top
    ax.text(score, cv + 1, f'{cv}%', 
            ha='center', va='bottom', fontsize=13, fontweight='bold', color='#2C3E50')
    # Sample size in middle of bar
    ax.text(score, cv/2, f'n={count}', 
            ha='center', va='center', fontsize=11, color='white', fontweight='bold')

# Labels and title
ax.set_xlabel('Infrastructure Score (Number of Key Indicators Present)', fontsize=13, fontweight='bold', color='#003D82')
ax.set_ylabel('Volatility (Coefficient of Variation %)', fontsize=13, fontweight='bold', color='#003D82')
ax.set_title('Infrastructure Predicts Stability in Maternal Health Outcomes\n(Lower Volatility = More Predictable Performance for Value-Based Care)', 
             fontsize=14, fontweight='bold', color='#003D82', pad=20)

# X-axis labels
ax.set_xticks(infrastructure_scores)
ax.set_xticklabels(['0\n(No\nIndicators)', 
                     '1\n(1 Indicator)', 
                     '2\n(2 Indicators)', 
                     '3\n(3 Indicators)', 
                     '4\n(All 4\nIndicators)'],
                   fontsize=10)

# Y-axis
ax.set_ylim(0, 40)
ax.set_yticks(np.arange(0, 41, 5))
ax.yaxis.grid(True, linestyle='--', alpha=0.3, color='gray')

# Add annotation showing the key finding
ax.annotate('54% Lower Volatility\n(4.7x More Stable)', 
            xy=(4, 15.2), xytext=(2.3, 8),
            arrowprops=dict(arrowstyle='->', color='#F4A03F', lw=3),
            fontsize=12, fontweight='bold', color='#F4A03F',
            bbox=dict(boxstyle='round,pad=0.7', facecolor='white', 
                     edgecolor='#F4A03F', linewidth=2))

# Clean styling
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#2C3E50')
ax.spines['bottom'].set_color('#2C3E50')

# Adjust layout FIRST to make room for footnote
plt.tight_layout()
plt.subplots_adjust(bottom=0.22)  # More space at bottom for footnote

# Add note about indicators AFTER adjusting layout - NOW LOWER
fig.text(0.5, 0.06, 
         'Four Infrastructure Indicators: (1) Medicaid Expansion, (2) Medicaid Doula Coverage, (3) Paid Family Leave, (4) Established MMRC (10+ years)',
         ha='center', fontsize=9, style='italic', color='#2C3E50')

fig.text(0.5, 0.02, 
         'States scored 0-4 based on number of indicators present. Indicators map to OE-3PI Framework\'s three pillars: Process, People, Technology.',
         ha='center', fontsize=9, style='italic', color='#2C3E50')

# Save high-resolution version for policy brief
plt.savefig('volatility_by_infrastructure_chart.png', 
            dpi=300, bbox_inches='tight', facecolor='white')

print("âœ… Chart saved as: volatility_by_infrastructure_chart.png")
plt.show()
```

```python
"""
INFRASTRUCTURE SCORE TABLE - STYLED VERSION
Professional table showing MMR by infrastructure score

CORRECTED: Shows 4 indicators that map to 3 pillars
- Medicaid Expansion (Process pillar)
- Doula Coverage (People pillar)
- Paid Family Leave (People pillar)
- Established MMRC (Technology pillar)
"""

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from your framework validation
data = {
    'Infrastructure Score': ['Score 4\n(All 4 Indicators)', 
                             'Score 3\n(3 Indicators)', 
                             'Score 2\n(2 Indicators)', 
                             'Score 1\n(1 Indicator)', 
                             'Score 0\n(No Indicators)'],
    'Number of States': [7, 6, 6, 5, 5],
    'Average MMR\n(per 100K)': [20.9, 28.1, 29.4, 36.0, 39.2],
    'MMR Range': [7.3, 24.1, 22.4, 27.0, 34.2],
    'Stability': ['Highly Stable', 'Moderate', 'Moderate', 'Volatile', 'Highly Volatile']
}

df = pd.DataFrame(data)

# Create figure with extra space for footnote
fig, ax = plt.subplots(figsize=(14, 6.5), facecolor='white')
ax.axis('tight')
ax.axis('off')

# Create table
table = ax.table(cellText=df.values, colLabels=df.columns,
                 cellLoc='center', loc='center',
                 colWidths=[0.25, 0.15, 0.18, 0.15, 0.18])

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2.5)

# Color code by infrastructure score
colors = ['#32CD32', '#FFD700', '#FFA500', '#FF6347', '#DC143C']

# Header styling
for i, col in enumerate(df.columns):
    cell = table[(0, i)]
    cell.set_facecolor('#003D82')
    cell.set_text_props(weight='bold', color='white', fontsize=12)
    cell.set_edgecolor('#2C3E50')
    cell.set_linewidth(2)

# Row styling
for i in range(len(df)):
    # Color code the infrastructure score column
    cell = table[(i+1, 0)]
    cell.set_facecolor(colors[i])
    cell.set_text_props(weight='bold', color='white', fontsize=11)
    cell.set_edgecolor('#2C3E50')
    cell.set_linewidth(1.5)
    
    # Style other cells
    for j in range(1, len(df.columns)):
        cell = table[(i+1, j)]
        cell.set_facecolor('#F8F9FA')
        cell.set_edgecolor('#2C3E50')
        cell.set_linewidth(1)
        
        # Bold the MMR values
        if j == 2:  # Average MMR column
            cell.set_text_props(weight='bold', fontsize=11)

# Add title
plt.title('Maternal Mortality by Infrastructure Score: States with More Indicators Show Lower, More Stable Outcomes',
          fontsize=14, fontweight='bold', color='#003D82', pad=20, loc='center')

# Adjust layout for footnote
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)

# Add caption/note - CORRECTED
fig.text(0.5, 0.08, 
         'Infrastructure scored 0-4 based on presence of four key indicators: (1) Medicaid Expansion, (2) Medicaid Doula Coverage,',
         ha='center', fontsize=9.5, style='italic', color='#2C3E50')

fig.text(0.5, 0.04, 
         '(3) Paid Family Leave, (4) Established MMRC (10+ years). These indicators map to the OE-3PI Framework\'s three pillars: Process, People, Technology.',
         ha='center', fontsize=9.5, style='italic', color='#2C3E50')

plt.savefig('infrastructure_mmr_table.png', dpi=300, bbox_inches='tight', facecolor='white')
print("âœ… Table saved: infrastructure_mmr_table.png")
plt.show()
```

```python
"""
INFRASTRUCTURE SCORE COMPARISON - BAR CHART VERSION
Shows MMR by infrastructure score with visual bars

CORRECTED: 4 indicators across 3 pillars
"""

import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Score 4\n(All 4 Indicators)', 
              'Score 3\n(3 Indicators)', 
              'Score 2\n(2 Indicators)', 
              'Score 1\n(1 Indicator)', 
              'Score 0\n(No Indicators)']
mmr_values = [20.9, 28.1, 29.4, 36.0, 39.2]
state_counts = [7, 6, 6, 5, 5]
colors = ['#32CD32', '#FFD700', '#FFA500', '#FF6347', '#DC143C']

# Create figure with space for footnote
fig, ax = plt.subplots(figsize=(12, 8), facecolor='white')

# Horizontal bars
y_pos = np.arange(len(categories))
bars = ax.barh(y_pos, mmr_values, color=colors, edgecolor='#2C3E50', linewidth=2, alpha=0.9)

# Add value labels on bars
for i, (mmr, count) in enumerate(zip(mmr_values, state_counts)):
    ax.text(mmr + 1, i, f'MMR: {mmr}  |  {count} states', 
            va='center', fontsize=12, fontweight='bold', color='#2C3E50')

# Add reference line at national average
national_avg = 18.2
ax.axvline(x=national_avg, color='#003D82', linestyle='--', linewidth=2.5, 
           alpha=0.7, label=f'National Average (2023): {national_avg}')

# Styling
ax.set_yticks(y_pos)
ax.set_yticklabels(categories, fontsize=12, fontweight='bold')
ax.set_xlabel('Maternal Mortality Rate (per 100,000 live births)', fontsize=13, fontweight='bold', color='#003D82')
ax.set_title('Infrastructure Predicts Maternal Mortality Outcomes\nStates with All 4 Indicators Show 47% Lower Mortality', 
             fontsize=15, fontweight='bold', color='#003D82', pad=20)

ax.set_xlim(0, 50)
ax.xaxis.grid(True, alpha=0.3, linestyle='--')
ax.legend(loc='lower right', fontsize=11, framealpha=0.9)

# Invert y-axis so best performers are on top
ax.invert_yaxis()

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Adjust for footnote
plt.tight_layout()
plt.subplots_adjust(bottom=0.12)

# Add footnote - CORRECTED
fig.text(0.5, 0.04, 
         'Four Infrastructure Indicators: (1) Medicaid Expansion (Process), (2) Doula Coverage (People), (3) Paid Family Leave (People), (4) Established MMRC (Technology)',
         ha='center', fontsize=9, style='italic', color='#2C3E50')

plt.savefig('infrastructure_mmr_horizontal_bars.png', dpi=300, bbox_inches='tight', facecolor='white')
print("âœ… Chart saved: infrastructure_mmr_horizontal_bars.png")
plt.show()
```
