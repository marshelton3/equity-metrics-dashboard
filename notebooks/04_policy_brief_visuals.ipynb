```python
"""
NATIONAL MATERNAL MORTALITY TREND 2018-2023
Shows overall national improvement following Medicaid extension
For Policy Brief - FIXED TEXT OVERLAP
"""

import matplotlib.pyplot as plt
import numpy as np

# National MMR data (all races combined)
years = ['2018', '2019', '2020', '2021', '2022', '2023']
mmr_national = [17.4, 20.1, 23.8, 32.9, 22.3, 18.2]

# Create figure
fig, ax = plt.subplots(figsize=(12, 7.5), facecolor='white')

# Main trend line
line = ax.plot(years, mmr_national, 
               marker='o', markersize=12, linewidth=3.5, 
               color='#003D82', markerfacecolor='#F4A03F', 
               markeredgewidth=2.5, markeredgecolor='#003D82',
               label='National MMR')

# Highlight key points
ax.scatter([3], [32.9], s=500, color='#DC143C', alpha=0.25, zorder=5)
ax.annotate('COVID Peak\n32.9 per 100K', 
            xy=(3, 32.9), xytext=(3, 37),
            ha='center', fontsize=11, fontweight='bold', color='#DC143C',
            arrowprops=dict(arrowstyle='->', color='#DC143C', lw=2))

ax.scatter([5], [18.2], s=500, color='#32CD32', alpha=0.25, zorder=5)
ax.annotate('Post-Extension\n18.2 per 100K', 
            xy=(5, 18.2), xytext=(5, 13),
            ha='center', fontsize=11, fontweight='bold', color='#32CD32',
            arrowprops=dict(arrowstyle='->', color='#32CD32', lw=2))

# Add improvement arrow
ax.annotate('', 
            xy=(5, 18.2), xycoords='data',
            xytext=(3, 32.9), textcoords='data',
            arrowprops=dict(arrowstyle='<->', color='#F4A03F', lw=3))

ax.text(4, 26, '45% Improvement\n(14.7 point decline)', 
        ha='center', fontsize=13, fontweight='bold', color='#F4A03F',
        bbox=dict(boxstyle='round,pad=0.7', facecolor='white', 
                 edgecolor='#F4A03F', linewidth=2))

# Add policy implementation line
ax.axvline(x=3.5, color='#008C95', linestyle='--', linewidth=3, alpha=0.7, 
           label='Medicaid Postpartum\nExtension (April 2022)')

# Add value labels on all points
for i, (year, mmr) in enumerate(zip(years, mmr_national)):
    if i not in [3, 5]:  # Skip the annotated points
        # SPECIAL HANDLING FOR 22.3 (index 4) - Move it down more
        if i == 4:
            ax.text(i, mmr - 2, f'{mmr}',  # MOVED DOWN (mmr - 2 instead of mmr + 1.5)
                    ha='center', va='top', fontsize=11, fontweight='bold', color='#2C3E50')
        else:
            ax.text(i, mmr + 1.5, f'{mmr}', 
                    ha='center', va='bottom', fontsize=11, fontweight='bold', color='#2C3E50')

# Labels and styling
ax.set_xlabel('Year', fontsize=14, fontweight='bold', color='#003D82')
ax.set_ylabel('Maternal Mortality Rate (per 100,000 live births)', fontsize=14, fontweight='bold', color='#003D82')
ax.set_title('National Maternal Mortality Trend: Dramatic Improvement Following Medicaid Postpartum Extension\nApproximately 550 Fewer Maternal Deaths Annually', 
             fontsize=15, fontweight='bold', color='#003D82', pad=20)

ax.set_ylim(0, 40)
ax.set_yticks(np.arange(0, 41, 5))
ax.yaxis.grid(True, alpha=0.3, linestyle='--', color='gray')
ax.legend(loc='upper left', fontsize=11, framealpha=0.9)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#2C3E50')
ax.spines['bottom'].set_color('#2C3E50')

# Spacing
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)

# Add data source note
fig.text(0.5, 0.06, 
         'Source: CDC WONDER Database, 2018-2023. Maternal mortality defined as deaths during pregnancy or within 42 days of pregnancy termination.',
         ha='center', fontsize=9, style='italic', color='#2C3E50')

plt.savefig('national_mmr_trend_2018_2023.png', dpi=300, bbox_inches='tight', facecolor='white')
print("âœ… Chart saved: national_mmr_trend_2018_2023.png")
plt.show()
```
